```python
# Dependencies
import pandas as pd
from geopy import distance
from ast import literal_eval
```

```python
# Store filepath in a variable
file_one = "202206-citbike-tripdata.csv"
```

```python
# Read our Data file with the pandas library
file_one_df = pd.read_csv(file_one)
```

```python
file_sample = file_one_df.sample(10000)
```

```python
file_sample = file_sample.dropna()
```

```python
file_sample.head()
```

```python
def pretty_coords_start (row):
    coord = '(' + str(row['start_lat']) + ',' + str(row['start_lng']) + ')'
    return coord

def pretty_coords_end (row):
    coord = '(' + str(row['end_lat']) + ',' + str(row['end_lng']) + ')'
    return coord
```

```python
def coords_distance (row):
    
    coord = distance.distance(literal_eval(row['start_loc_pretty']),literal_eval(row['end_loc_pretty'])).miles
    return coord
```

```python
file_sample.apply (lambda row: coords_distance(row), axis=1)
```

```python
#lets create grouped by station version for cool charting
file_sample_grouped_station = file_sample.groupby(['start_station_name']).mean()
# file_sample_grouped_station = file_sample_grouped_station[['loc_distance']]
file_sample_grouped_station
```

```python
file_sample_grouped_station_sum = file_sample.groupby(['start_station_name']).sum()
file_sample_grouped_station_sum = file_sample_grouped_station_sum[['loc_distance']]
file_sample_grouped_station_sum
```

```python
#get a count
file_sample_count = file_sample.groupby(['start_station_name']).count()
file_sample_count = file_sample_count['ride_id']
file_sample_count.columns = ['count']
file_sample_count
```

```python
#concat
file_sample_grouped_station['count'] = file_sample_count
file_sample_grouped_station
```

```python
#concat
file_sample_grouped_station['sum'] = file_sample_grouped_station_sum
file_sample_grouped_station
```

```python
file_sample_grouped_station = file_sample_grouped_station.rename(columns={'loc_distance': 'distance_mean', 'count': 'ride_count',
'sum': 'distance_sum'})
file_sample_grouped_station
```

```python
file_sample_grouped_station = file_sample_grouped_station.reset_index()
file_sample_grouped_station
```
